# Using the Yelp API (Practice)

```python
import pandas as pd
```

```python
#Collect the results of a search for SLO pizza
results = yelp_api.search_query(location='San Luis Obispo, CA',
                                       term='Pizza')
print(type(results))
results.keys()
```

```python
#Convert the businesses into a dataframe
df = pd.DataFrame(results['businesses'])
df.head()
```

```python
#Total businesses 
results['total']
```

```python
#Quantity of businesses returned from the search due to pagination
len(results['businesses'])
```

```python
#Add an offset to get the next page of search results
results2 = yelp_api.search_query(location='San Luis Obispo, CA',
                                       term='Pizza',
                                offset = 20)
print(type(results2))
results2.keys()
```

```python
#Convert the second search to a dataframe
df2 = pd.DataFrame(results2['businesses'])
df2.head()
```

```python
#Concatenate both df's into one called slo_pizza.
slo_pizza = pd.concat([df, df2],
                     ignore_index=True)
display(slo_pizza.head(), slo_pizza.tail())
```

```python
#Save the results as a json file.
pizza_json = slo_pizza.to_json(orient='records')
```
